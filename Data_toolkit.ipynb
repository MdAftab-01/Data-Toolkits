### **2. How does broadcasting work in NumPy?**

**Broadcasting** allows NumPy to perform operations on arrays of **different shapes** without explicitly copying data.

âœ… **Example:**

```python
import numpy as np
a = np.array([1, 2, 3])
b = 2
print(a + b)  # Output: [3, 4, 5]
```

Here, the scalar `b` is **broadcast** to match the shape of `a`.

âœ… **Another example (2D + 1D):**

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
print(A + B)  # B is broadcast to each row of A
```

ðŸ§  **Broadcasting Rules**:

* Start from the **trailing dimensions** and compare shapes.
* Dimensions must be **equal** or **1**.

### **21. Why is NumPyâ€™s array slicing faster than Pythonâ€™s list slicing?**

âœ… **Reason:**
NumPy arrays are **stored in contiguous memory blocks** and implemented in **C**, allowing faster data access and manipulation.

| Feature           | **NumPy Array**                  | **Python List**                  |
| ----------------- | -------------------------------- | -------------------------------- |
| **Memory layout** | Contiguous, fixed-type (C-style) | Scattered, variable-type objects |
| **Speed**         | Very fast (C-level operations)   | Slower (Python object overhead)  |
| **Overhead**      | Minimal                          | High per element                 |

ðŸ§  **In short:**
NumPy slicing returns a **view**, not a copy â€” making it **faster and more memory-efficient** than slicing regular Python lists.
